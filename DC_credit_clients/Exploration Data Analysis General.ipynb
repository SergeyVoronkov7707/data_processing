```python
# преобразование категориальных в числовые
def cat_to_num(data):
    count = 1
    if pd.Series == type(data):
        res_dict = {}
        n_un = data.unique()
        for n, i in enumerate(n_un, 1):
                try:
                    res_dict[i] = n/(len(n_un) * 10)
                    count += 1
                except:
                    pass
                
#         print(res_dict)
        data = data.map(res_dict)
        return data
    else:
        d_c = []
        for sign in data.dtypes.items():
            if sign[1] == "object":
                res_dict = {}
                n_un = data[sign[0]].unique()
                for n, i in enumerate(n_un, 1):
                    try:
                        res_dict[i] = n/(len(n_un) * 10)
                        count += 1
                    except:
                        pass
#                 print(res_dict)
                data[sign[0]] = data[sign[0]].map(res_dict)
        return data   
```

```python
def cor(data, target):
    correlation = data.corr()
    corr_with_target = correlation[target].sort_values(ascending = False)
    print(corr_with_target)
    fig, axes = plt.subplots(figsize = (14,12))
    plt.title("Correlation of Numeric Features with target", y=1, size=16)
    sns.heatmap(correlation, square=True, vmax=0.8, cmap="viridis")
```

```python
# ф-я которая принимает фрейм с таргетом и фрейм с признаками, список признаков для получения новых признаков
# если type -> object преобразуем в числа и берем среднее по колонке относительно id сразу делаем merge с фреймом с таргетом
# если int, float берем среднее по колонке относительно id сразу делаем merge с фреймом с таргетом
def new_feature(data_1,data_2, colums_name: list,columns_leave:list, target= -1):
    df = pd.DataFrame()
    df[columns_leave] = data_1[columns_leave]
    data = data_2.copy()
    if type(colums_name) != list:
        colums_name = list(colums_name)
    
    for col in tqdm(colums_name):
        if 'O' ==data_2[col].dtype:
            data[col] = cat_to_num(data[col])
            apps_groupby = data.groupby("APPLICATION_NUMBER", as_index=False)
            previous_app_counts = apps_groupby[col].mean()
            new_name = "PREV_" + str(col) +"_MEAN"
            previous_app_counts = previous_app_counts.rename(columns={
                    str(col): new_name
                    })
            data_1 = data_1.merge(
            previous_app_counts, how="inner", on="APPLICATION_NUMBER"
                    )
            df[new_name] = data_1[new_name].fillna(-100)
        else:
            apps_groupby = data_2.groupby("APPLICATION_NUMBER", as_index=False)
            previous_app_counts = apps_groupby[col].mean()
            new_name = "PREV_" + str(col) +"_MEAN"
            previous_app_counts = previous_app_counts.rename(columns={
                    str(col): new_name
                    })
            data_1 = data.merge(
            previous_app_counts, how="inner", on="APPLICATION_NUMBER"
                    )
            df[str(col)] = data_1[str(col)].fillna(-100)
#             df[new_name] = data_1[new_name].fillna(-100)
    if target != -1:
        cor(df, target)
    return df
```

```python
def merge_data(data, merge_data):
    data_1 = data.merge(
            merge_data, how="left", on="APPLICATION_NUMBER"
                    )
    return data_1
```

```python
def calculate_permutation_importance(estimator, 
                                     metric: callable,
                                     x_valid: pd.DataFrame,
                                     y_valid: pd.Series) -> pd.Series:
    """
    Расчет пермутированной важности признаков.
    """
    scores = {}
    y_pred = estimator.predict(xgb.DMatrix(x_valid))
    base_score = metric(y_valid, y_pred)

    for feature in tqdm(x_valid.columns):
        x_valid_copy = x_valid.copy()
        x_valid_copy[feature] = np.random.permutation(x_valid_copy[feature])

        y_pred = estimator.predict(xgb.DMatrix(x_valid_copy))
        score = metric(y_valid, y_pred)
        scores[feature] = base_score - score

    scores = pd.Series(scores)
    scores = scores.sort_values(ascending=False)

    return scores
```

```python
"""
ф-я определяет пермутированную важность признаков
"""
def perm_connection(data, target, list_drop, metric=roc_auc_score):
    x_train, x_valid = train_test_split(
    data.drop(list_drop, axis=1), train_size=0.7, random_state=1
        )
    y_train, y_valid = train_test_split(
    data[target], train_size=0.7, random_state=1
        )
    model = train_model_perm(x_train, x_valid, y_train, y_valid)
    perm_importance = calculate_permutation_importance(
    estimator=model, metric=metric, x_valid=x_valid, y_valid=y_valid
        )
    return perm_importance
```

```python
def train_model_perm(x_train, x_valid, y_train, y_valid):
    xgb_params = {
    "booster": "gbtree",
    "objective": "binary:logistic",
    "eval_metric": "auc",
    "learning_rate": 0.1,
    "n_estimators": 1000,
    "nthread": 6,
    "seed": 27
        }

    dtrain = xgb.DMatrix(x_train, y_train)
    dvalid = xgb.DMatrix(x_valid, y_valid)

    model = xgb.train(
    dtrain=dtrain,
    params=xgb_params,
    num_boost_round=500,
    evals=[(dtrain, "train"), (dvalid, "valid")],
    early_stopping_rounds=50,
    verbose_eval=10,
        )
    return model
```

```python
"""
ф-я определяет нелинейную связь признаков, возвращает список из 15 наиболее важных
"""
def nonlianer_comun(data, target="TARGET")-> list:
    parameters = {"max_depth": 6, "n_estimators": 25, "random_state": 27, "n_jobs": 2}

    forest = RandomForestRegressor(**parameters)
    forest.fit(data.drop(target, axis=1), data[target])
    
    # numerical original
    top = []
    n_top = 15
    importances = forest.feature_importances_
    idx = np.argsort(importances)[::-1][0:n_top]
    feature_names = data.drop("TARGET", axis=1).columns

    plt.figure(figsize=(20, 5))
    sns.barplot(x=feature_names[idx], y=importances[idx], palette="viridis")
    top.append(feature_names[idx])
    plt.title("What are the top important features to start with?", size=14)
    plt.xticks(rotation=90)
    return top
```

```python
"""
ф-я преобразования тип обьект в числа
"""
def obj_to_num(data, columns=None):
    
    print(columns)
    train = pd.DataFrame()
    if columns == None:
        
        features =  [n  for n,i in data.dtypes.items() if i=="O"]
        col = data.columns
        col = [i for i in col if i not in features]
    else:
        if type(columns) != list:
            columns = list(columns)
        features = columns
        col = data[columns]
        col = [i for i in col if i not in features]
    encoder = OrdinalEncoder(
            handle_unknown='use_encoded_value', 
            unknown_value=-1
        ).fit(data[features])
    train[features] = encoder.transform(data[features])
    train[col] = data[col]
    return train
```

```python
# ф-я выводит кол-во пропусков кол-во и в процентах и сразу удаляет колонки где более 50% пропусков
# и если меньше 10% заменяет на медиану или моду(зависит от типа)

def chech_pass(data, percent=50, min_percent = 10):
    min_percent = min_percent
    percent = percent
    del_col = {}
    median_col = {}
    mode_col = {}
    print(data.shape)
# print(applications_history.isna().sum())
    for i, col in list(zip(data.isna().sum(),data.columns)):
        print("{}: кол-во пропусков: {}, процент: {} % ".format(col,i , round(i/data.shape[0] * 100), 1))
        if i/data.shape[0] * 100 > percent:
            data = data.drop(col, 1)
            del_col[col] = round(i/data.shape[0] * 100)
        elif i/data.shape[0] * 100 < min_percent and data[col].dtype != "O":
            data[col].fillna(data[col].median(), inplace = True)
            median_col[col] = round(i/data.shape[0] * 100)
        elif i/data.shape[0] * 100 < min_percent and data[col].dtype == "O":
            data[col].fillna(data[col].mode(), inplace = True)
            mode_col[col] = round(i/data.shape[0] * 100)
    if len(del_col) == 0:
        print(f"В данных нету пропусков более {percent} %")
    for d, n in del_col.items():
        print("Удаленные колонки: {} c {}%".format(d, n))
    for d, n in median_col.items():
        print("В колонке: {} значение заменнино на медиану с пропуском: {}%".format(d, n))
    for d, n in mode_col.items():
        print("В колонке: {} значение заменнино на моду с пропуском: {}%".format(d, n))
    return data
```

```python
"""
ф-я удаляет все, что выше 99 перцентиля и ниже 1 перцентиля
"""
def del_perc_mm(data, columns):
    out = pd.DataFrame(columns= data.columns)
    for col in columns:
        LEFT_BOUND, RIGHT_BOUND = np.percentile(data[col], q=1), np.percentile(data[col], q=99)
        x_winsorized = np.clip(data[col], LEFT_BOUND, RIGHT_BOUND)
        assert x_winsorized.shape[0] == data[col].shape[0]
        out[col] = x_winsorized
    return out
```
